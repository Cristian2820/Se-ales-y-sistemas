```python
%matplotlib inline
import sympy as sym
sym.init_printing()

t, tau = sym.symbols('t tau', real=True)#variables simb贸licas
```

```python
class rect(sym.Function): #funci贸n pulso rectangular
    @classmethod
    def eval(cls, arg):
        return sym.Heaviside(arg + sym.S.Half) - sym.Heaviside(arg - sym.S.Half)
```

```python
def plot_signals(x_axis, x, h, ylabel, xlabel):#funci贸n para graficar convoluci贸n
    p1 = sym.plot(x, (x_axis, -10, 10), show=False, line_color='b', ylabel=ylabel, xlabel=xlabel)
    p2 = sym.plot(h, (x_axis, -10, 10), show=False, line_color='r')
    p1.extend(p2)
    p1.show()
```

```python
h = sym.exp(-t) * sym.Heaviside(t)
x = rect(t - 1/2)

plot_signals(t, x, h, r'$h(t)$, $x(t)$', r'$t$')
```

```python
h1 = h.subs(t, tau)
x1 = x.subs(t, tau)

plot_signals(tau, x1, h1, r'$h(\tau)$, $x(\tau)$', r'$\tau$')
```

```python
h2 = h1.subs(tau, -tau)

plot_signals(tau, x1, h2, r'$h(-\tau)$, $x(\tau)$', r'$\tau$')
```

```python
h3 = h2.subs(tau, tau-t)

plot_signals(tau, x1, h3.subs(t, -2), r'$h(t-\tau)$, $x(\tau)$', r'$\tau$')
```

```python
h4 = h3.subs(tau, tau-t)
plot_signals(tau, x1, h3.subs(t, -1), r'$h(t-\tau)$, $x(\tau)$', r'$\tau$')
```

```python
h5 = h4.subs(tau, tau-t)
plot_signals(tau, x1, h3.subs(t, 0.5), r'$h(t-\tau)$, $x(\tau)$', r'$\tau$')
```

```python
h6 = h5.subs(tau, tau-t)
plot_signals(tau, x1, h3.subs(t, 6), r'$h(t-\tau)$, $x(\tau)$', r'$\tau$')
```

```python
h7 = h6.subs(tau, tau-t)
plot_signals(tau, x1, h3.subs(t, 8), r'$h(t-\tau)$, $x(\tau)$', r'$\tau$')
```
