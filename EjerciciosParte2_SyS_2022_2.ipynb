## 1. Representaci√≥n generalizada y serie de Fourier

**1.6 ¬øComo se puede simplificar el c√°lculo de la serie trigonom√©trica de Fourier para se√±ales con simetr√≠a de media onda y de cuarto de onda?. En dichos casos, ¬øcomo se pueden calcular los coeficientes $(a_n)$ y $(b_n)$ con n ‚àà N?.**

**1.Simetr√≠a de Media Onda**

Una se√±al $ùëì(ùë°)$ tiene simetr√≠a de media onda si cumple con:

$$ f(t + T/2) = -f(t) $$

donde $ùëá$ es el per√≠odo de la se√±al.

**Coeficientes de Fourier**

Para una se√±al con simetr√≠a de media onda, los coeficientes $a_n$ (coeficientes de coseno) de la serie de Fourier son cero. Esto se debe a que $f(t)$ es una funci√≥n impar, y las funciones coseno en la serie de Fourier representan componentes pares.

La serie de Fourier para una se√±al con simetr√≠a de media onda se simplifica a:

$$ f(t) = \sum_{n=1}^{\infty} b_n \sin\left(\frac{2 \pi n t}{T}\right) $$

donde los coeficientes $b_n$ se calculan como:

$$ b_n = \frac{2}{T} \int_{0}^{T} f(t) \sin\left(\frac{2 \pi n t}{T}\right) \, dt $$


**2.Simetr√≠a de Cuarto de Onda:**

Una se√±al $ùëì(ùë°)$ tiene simetr√≠a de cuarto de onda si cumple con:

$$ f(t + T/4) = -f(t) $$

Esto implica que la se√±al es impar con respecto a $ùëá/4$. Para una se√±al con simetr√≠a de cuarto de onda, los coeficientes $b_n$ (coeficientes de seno) de la serie de Fourier son cero. Esto es porque la se√±al es par con respecto a $ùëá/4$, y las funciones seno en la serie de Fourier representan componentes impares.

La serie de Fourier para una se√±al con simetr√≠a de cuarto de onda se simplifica a:

$$ f(t) = a_0 + \sum_{n=1}^{\infty} a_n \cos\left(\frac{2 \pi n t}{T}\right) $$

donde los coeficientes $a_n$ se calculan como:

$$ a_n = \frac{2}{T} \int_{0}^{T} f(t) \cos\left(\frac{2 \pi n t}{T}\right) \, dt $$

**Serie de Fourier**

**1.Serie de Fourier Trigonom√©trica:**
*   Descripci√≥n: Representa una funci√≥n peri√≥dica como una suma infinita de senos y cosenos. Se expresa en t√©rminos de coeficientes $a_n$ y $b_n$ para cada arm√≥nico.
*   F√≥rmula:
$$ f(t) = a_0 + \sum_{n=1}^{\infty} \left(a_n \cos\left(\frac{2 \pi n t}{T}\right) + b_n \sin\left(\frac{2 \pi n t}{T}\right)\right) $$


**2.Serie de Fourier Exponencial:**
*   Descripci√≥n: Utiliza funciones exponenciales complejas en lugar de senos y cosenos. Es √∫til para representar se√±ales peri√≥dicas y se basa en la forma compleja de la serie de Fourier.
*   F√≥rmula:
$$ f(t) = \sum_{n=-\infty}^{\infty} c_n e^{j \frac{2 \pi n t}{T}} $$
*   Coeficientes:
$$ c_n = \frac{1}{T} \int_{0}^{T} f(t) e^{-j \frac{2 \pi n t}{T}} \, dt $$

**3.Serie de Fourier Compacta:**
*   Descripci√≥n: Tambi√©n conocida como la serie de Fourier de Fourier, usa una representaci√≥n compacta de los coeficientes en t√©rminos de senos y cosenos o exponenciales, con una notaci√≥n simplificada.
*   F√≥rmula: Igual a la serie exponencial, pero frecuentemente simplificada en la notaci√≥n.

**Transformadas de Fourier**

**1.Transformada de Fourier (TF):**
*   Descripci√≥n: Generaliza la serie de Fourier para funciones no peri√≥dicas. Convierte una se√±al en el dominio del tiempo en una representaci√≥n en el dominio de la frecuencia continua.
*   Formula:
$$ F(\omega) = \int_{-\infty}^{\infty} f(t) e^{-j \omega t} \, dt $$

**2.Transformada de Fourier en Tiempo Discreto (DTFT):**
*  Descripci√≥n: Generaliza la TF para se√±ales discretas en el tiempo, resultando en una representaci√≥n continua en frecuencia.
*   F√≥rmula:
$$ X(\omega) = \sum_{n=-\infty}^{\infty} x[n] e^{-j \omega n} $$
*  Nota: La DTFT es una funci√≥n continua de $ùúî$, a diferencia de la DFT que es discreta.

**3.ransformada Discreta de Fourier (DFT):**
*  Descripci√≥n: Se utiliza para calcular la representaci√≥n de frecuencia de se√±ales discretas en tiempo y en frecuencia. Es una versi√≥n discreta y finita de la transformada de Fourier.
*   F√≥rmula:
$$ X[k] = \sum_{n=0}^{N-1} x[n] e^{-j \frac{2 \pi k n}{N}} $$
*  Nota: La DFT convierte una secuencia de longitud $ùëÅ$ en $ùëÅ$ componentes en frecuencia.

**Algoritmo de Transformada R√°pida de Fourier (FFT)**

**1.Descripci√≥n:**
*   FFT es un algoritmo eficiente para calcular la DFT y su inversa (IDFT). Reduce el costo computacional significativamente en comparaci√≥n con el c√°lculo directo de la DFT.
*   Funci√≥n: La FFT divide el c√°lculo de la DFT en sub-problemas m√°s peque√±os, utilizando un enfoque de divide y vencer√°s.

**2.Detalle del Algoritmo:**
*   Divide y Vencer√°s: La FFT divide la secuencia en dos mitades, resuelve la DFT de cada mitad recursivamente y combina los resultados.
*   Ejemplo de Implementaci√≥n: La FFT de Cooley-Tukey es uno de los algoritmos m√°s conocidos, y tiene una complejidad computacional de
$ùëÇ(ùëÅlogùëÅ)$, donde $ùëÅ$ es el n√∫mero de puntos en la secuencia.

**3.Costo Computacional:**
*   DFT: La complejidad computacional es $ùëÇ(ùëÅ^2)$, ya que cada punto de la DFT requiere la suma de $ùëÅ$ t√©rminos.
*   FFT: La complejidad es $ùëÇ(ùëÅlogùëÅ)$, lo que es mucho m√°s eficiente para $ùëÅ$ grandes.

**Distorsi√≥n Total de Arm√≥nicos (THD):** es una medida de la distorsi√≥n en una se√±al el√©ctrica causada por la presencia de arm√≥nicos, que son frecuencias m√∫ltiples de la frecuencia fundamental. THD se expresa como un porcentaje y se calcula usando la f√≥rmula:

$$ \text{THD} = \frac{\sqrt{\sum_{n=2}^{N} V_n^2}}{V_1} \times 100\% $$

donde:

*   $( V_n )$ es la magnitud del \( n \)-√©simo arm√≥nico,
*   $( V_1 )$ es la magnitud de la frecuencia fundamental
*   \( N \) es el n√∫mero total de arm√≥nicos considerados.






**Factor de potencia:** es una medida que indica qu√© tan eficientemente se est√° utilizando la energ√≠a en un circuito el√©ctrico. Se define como el coseno del √°ngulo de fase (ùúô) entre la corriente y el voltaje:

$$ \text{PF} = \cos(\phi) $$

En un circuito con distorsi√≥n arm√≥nica, el factor de potencia puede verse afectado y se puede calcular a partir del THD utilizando la relaci√≥n:

$$ \text{PF}_{\text{dist}} = \frac{\text{PF}}{\sqrt{1 + \text{THD}^2}} $$

**C√°lculo del THD la FFT:** Para calcular el THD a partir de la Transformada R√°pida de Fourier (FFT), se siguen estos pasos:

1.   Realizar la FFT de la se√±al para obtener el espectro de frecuencia.
2.   Identificar las magnitudes de los componentes arm√≥nicos y la frecuencia fundamental.
3.   Calcular el THD usando la f√≥rmula mencionada anteriormente, donde las magnitudes de los arm√≥nicos se obtienen de la FFT.

**Carga Resistiva**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Par√°metros de la se√±al
fs = 1000  # Frecuencia de muestreo en Hz
T = 1.0 / fs  # Intervalo de muestreo
L = 1000  # N√∫mero de puntos
t = np.arange(0, L) * T  # Vector de tiempo

# Par√°metros del rectificador y carga resistiva
f = 50  # Frecuencia de la se√±al en Hz
V_in = 10  # Amplitud de la se√±al en V
R = 10  # Resistencia en Œ©

# Se√±al de salida del rectificador de onda completa
signal = V_in * np.abs(np.sin(2 * np.pi * f * t))

# Calcular FFT
yf = fft(signal)
xf = fftfreq(L, T)[:L // 2]  # Frecuencias

# Obtener magnitudes de la FFT
magnitudes = 2.0 / L * np.abs(yf[:L // 2])

# Calcular THD
V1 = magnitudes[1]  # Magnitud de la frecuencia fundamental (primer arm√≥nico)
Vn = np.sqrt(np.sum(magnitudes[2:]**2))  # Magnitud de los arm√≥nicos
THD = (Vn / V1) * 100

# Calcular el factor de potencia ideal
PF = 1.0  # Idealmente, para carga resistiva sin distorsi√≥n

# Calcular el factor de potencia distorsionado
PF_dist = PF / np.sqrt(1 + (THD / 100)**2)

# Graficar el espectro de frecuencia
plt.figure()
plt.plot(xf, magnitudes)
plt.title('Espectro de Frecuencia para Carga Resistiva')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud')
plt.grid()
plt.show()

print(f"THD: {THD:.2f}%")
print(f"PF (distorsionado): {PF_dist:.2f}")
```

**Carga RC**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Par√°metros de la se√±al
fs = 1000  # Frecuencia de muestreo en Hz
T = 1.0 / fs  # Intervalo de muestreo
L = 1000  # N√∫mero de puntos
t = np.arange(0, L) * T  # Vector de tiempo

# Par√°metros del rectificador y carga RC en serie
f = 50  # Frecuencia de la se√±al en Hz
V_in = 10  # Amplitud de la se√±al en V
R = 10  # Resistencia en Œ©
C = 1e-6  # Capacitancia en F (1 ŒºF)

# Se√±al de salida del rectificador de onda completa
signal = V_in * np.abs(np.sin(2 * np.pi * f * t))

# Calcular FFT
yf = fft(signal)
xf = fftfreq(L, T)[:L // 2]  # Frecuencias

# Obtener magnitudes de la FFT
magnitudes = 2.0 / L * np.abs(yf[:L // 2])

# Calcular THD
V1 = magnitudes[1]  # Magnitud de la frecuencia fundamental (primer arm√≥nico)
Vn = np.sqrt(np.sum(magnitudes[2:]**2))  # Magnitud de los arm√≥nicos
THD = (Vn / V1) * 100

# Calcular el factor de potencia ideal para carga RC
# Para simplicidad, asumimos que el factor de potencia ideal es el mismo para todos
PF = np.cos(np.arctan(1 / (2 * np.pi * f * R * C)))

# Calcular el factor de potencia distorsionado
PF_dist = PF / np.sqrt(1 + (THD / 100)**2)

# Graficar el espectro de frecuencia
plt.figure()
plt.plot(xf, magnitudes)
plt.title('Espectro de Frecuencia para Carga RC en Serie')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud')
plt.grid()
plt.show()

print(f"THD: {THD:.2f}%")
print(f"PF (distorsionado): {PF_dist:.2f}")
```

**2.5 Aplicacion en comunicaciones - modulaci√≥n AM: Consulte en que consiste la modulaci√≥n por amplitud por detecci√≥n coherente y sus aplicaciones. Genere un ejemplo ilustrativo sobre Python en el que se grafique las senales en el tiempo y en frecuencia (utilizando la "rfft") para: senal mensaje tipo pulso rectangular y senal mensaje tipo coseno. El usuario podra definir el √≠ndice de modulacion de inter√©s.**

**Modulaci√≥n AM:** La amplitud de la se√±al portadora cambia de acuerdo con la se√±al de mensaje. La se√±al modulada en AM se puede expresar como:

$$ s(t) = [A_c + A_m \cdot m(t)] \cdot \cos(2 \pi f_c t) $$

donde:

*  $( A_c)$ es la amplitud de la portadora,
*  $( A_m)$ es la amplitud m√°xima de la se√±al de mensaje,
*  $( m(t))$ es la se√±al de mensaje normalizada,
*  $( f_c )$ es la frecuencia de la portadora.





**Detecci√≥n Coherente:** Se utiliza un demodulador que multiplica la se√±al recibida por una copia sincronizada de la portadora y luego pasa el resultado por un filtro pasa-bajo para recuperar la se√±al de mensaje.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq

# Par√°metros de la se√±al
fs = 10000  # Frecuencia de muestreo en Hz
T = 1.0 / fs  # Intervalo de muestreo
L = 1000  # N√∫mero de puntos
t = np.arange(0, L) * T  # Vector de tiempo

# Par√°metros de modulaci√≥n
f_c = 1000  # Frecuencia de la portadora en Hz
A_c = 1  # Amplitud de la portadora
A_m = 0.5  # Amplitud m√°xima de la se√±al de mensaje

# Definir funci√≥n para graficar se√±al y espectro
def plot_am_modulation(message_signal, modulation_index, title):
    # Se√±al de mensaje
    m_t = message_signal(t)

    # Se√±al modulada AM
    s_t = (A_c + A_m * modulation_index * m_t) * np.cos(2 * np.pi * f_c * t)

    # Calcular FFT
    yf = rfft(s_t)
    xf = rfftfreq(L, T)  # Frecuencias

    # Graficar en el tiempo
    plt.figure(figsize=(14, 6))
    plt.subplot(2, 1, 1)
    plt.plot(t, s_t)
    plt.title(f'Se√±al AM Modulada: {title}')
    plt.xlabel('Tiempo [s]')
    plt.ylabel('Amplitud')
    plt.grid()

    # Graficar en frecuencia
    plt.subplot(2, 1, 2)
    plt.plot(xf, np.abs(yf))
    plt.title(f'Espectro de Frecuencia: {title}')
    plt.xlabel('Frecuencia [Hz]')
    plt.ylabel('Amplitud')
    plt.grid()

    plt.tight_layout()
    plt.show()

# Definir se√±ales de mensaje
def rectangular_pulse(t):
    return np.where((t % 0.02) < 0.01, 1, 0)  # Pulsos de 1 ms cada 20 ms

def cosine_wave(t):
    return np.cos(2 * np.pi * 50 * t)  # Se√±al cosenoidal de 50 Hz

# Modulaci√≥n para se√±al rectangular
modulation_index_rect = 1.0
plot_am_modulation(rectangular_pulse, modulation_index_rect, 'Se√±al de Mensaje Rectangular')

# Modulaci√≥n para se√±al cosenoidal
modulation_index_cos = 1.0
plot_am_modulation(cosine_wave, modulation_index_cos, 'Se√±al de Mensaje Cosenoidal')
```
