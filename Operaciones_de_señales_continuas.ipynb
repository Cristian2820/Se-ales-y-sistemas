Agregar otro escalón al principio de la señal escalera modificando el ejemplo anterior.

```python
import sympy as sym

# Definir la variable simbólica
t = sym.symbols('t', real=True)

# Definir la función rectangular
rect = sym.Heaviside(t + 0.5) - sym.Heaviside(t - 0.5)
sym.plot(rect, xlim=[-2, 2], ylabel=r'rect(t)')

# Agregar un escalón al principio
# Ajusta el primer coeficiente para que el nuevo escalón se vea correctamente al principio
x = 1/3 * rect.subs(t, t) + 2/3 * rect.subs(t, t - 0.5) + 1/3 * rect.subs(t, t - 1.5)

# Graficar la señal escalera modificada
sym.plot(x, (t, -1, 5), ylim=[-0.2, 1.2], ylabel='$x(t)$')
```

Modifique el ejemplo anterior para que la señal se comprima.

```python
import sympy as sym

# Definir la variable simbólica
t = sym.symbols('t', real=True)

# Definir la función rectangular
rect = sym.Heaviside(t + 0.5) - sym.Heaviside(t - 0.5)
x = 1/3 * rect.subs(t, t) + 2/3 * rect.subs(t, t - 0.5) + 1/3 * rect.subs(t, t - 1.5)

# Definir la variable de escala
a = sym.symbols('a', real=True)

# Crear la señal escalada
y = x.subs(t, a*t)

# Graficar la señal original
px = sym.plot(x, (t, -7, 7), ylim=[-0.2, 1.2], ylabel=r'$y(t)$', show=False, line_color='gray')

# Graficar la señal escalada con a=1/2
py = sym.plot(y.subs(a, 1/2), (t, -7, 7), ylim=[-0.2, 1.2], ylabel=r'$y(t)$', show=False)

# Extender los gráficos sobre una misma figura
px.extend(py)
px.show()
```

Modifique el ejemplo anterior para que la señal se escale y se invierta. ¿Qué factor de escala `a`da como resultado un       estiramiento/compresión en este contexto?  


```python
import sympy as sym

# Definir la variable simbólica
t = sym.symbols('t', real=True)

# Definir la función rectangular
rect = sym.Heaviside(t + 0.5) - sym.Heaviside(t - 0.5)
x = 1/3 * rect.subs(t, t) + 2/3 * rect.subs(t, t - 0.5) + 1/3 * rect.subs(t, t - 1.5)

# Definir la variable de escala
a = sym.symbols('a', real=True)

# Crear la señal escalada e invertida
y = x.subs(t, -a * t)

# Graficar la señal original
px = sym.plot(x, (t, -7, 7), ylim=[-0.2, 1.2], ylabel=r'$y(t)$', show=False, line_color='gray')

# Graficar la señal escalada e invertida con a=1/2
py = sym.plot(y.subs(a, 1/2), (t, -7, 7), ylim=[-0.2, 1.2], ylabel=r'$y(t)$', show=False, line_color='blue')

# Extender los gráficos sobre una misma figura
px.extend(py)
px.show()
```

*   Estiramiento: Si 0 < a < 1, la señal se estira.
*   Compresión: Si a > 1, la señal se comprime.
*   Inversión: Se logra con el signo negativo en el argumento.

En el contexto del código, si se desea un estiramiento de 2, el factor a debe ser 1/2. Si desea una compresión de 2, el factor a debe ser 2.

Para que valor de $\tau$, ¿la señal comienza en $t=0$?

\begin{align}
\text{rect}(t) &= \text{Heaviside}(t + 0.5) - \text{Heaviside}(t - 0.5)
\end{align}

\begin{align}
x(t) &= \frac{1}{3} \text{rect}(t) + \frac{2}{3} \text{rect}(t - 0.5) + \frac{1}{3} \text{rect}(t - 1.5)
\end{align}



\begin{align}
y(t) &= x(\tau - t)
\end{align}


\begin{align}
\tau + 0.5 &= 0 \\
\tau &= -0.5
\end{align}

\begin{align}
\tau &= 0.5
\end{align}

Para que la señal comience en \( t = 0 \), dado que la función rectangular comienza en \( -0.5 \), el valor de \( τ ) debe ser:

\begin{align}
\tau &= 0.5
\end{align}

```python
import sympy as sym

# Definir la variable simbólica
t = sym.symbols('t', real=True)

# Definir la función rectangular
rect = sym.Heaviside(t + 0.5) - sym.Heaviside(t - 0.5)
x = 1/3 * rect.subs(t, t) + 2/3 * rect.subs(t, t - 0.5) + 1/3 * rect.subs(t, t - 1.5)

# Definir la variable de escala
a = sym.symbols('a', real=True)

# Crear la señal comprimida (usando a > 1 para comprimir la señal)
compression_factor = 2  # Ajusta este valor para la compresión deseada
y = x.subs(t, t / compression_factor)

# Graficar la señal original
px = sym.plot(x, (t, -7, 7), ylim=[-0.2, 1.2], ylabel=r'$y(t)$', show=False, line_color='gray')

# Graficar la señal comprimida
py = sym.plot(y, (t, -7, 7), ylim=[-0.2, 1.2], ylabel=r'$y(t)$', show=False, line_color='blue')

# Extender los gráficos sobre una misma figura
px.extend(py)
px.show()
```

Realice un cambio temporal dividiendo en dos operaciones consecutivas: (i) Inversión temporal y (ii) Desplazamiento temporal.

Inversión Temporal:

\begin{align}
x(-t) &= \frac{1}{3} \text{rect}(-t) + \frac{2}{3} \text{rect}(-t - 0.5) + \frac{1}{3} \text{rect}(-t - 1.5)
\end{align}

Desplazamiento Temporal:

\begin{align}
x(-t - \tau) &= \frac{1}{3} \text{rect}(-t - \tau) + \frac{2}{3} \text{rect}(-t - \tau - 0.5) + \frac{1}{3} \text{rect}(-t - \tau - 1.5)
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función rectangular
def rect(t):
    return np.where(np.abs(t) <= 0.5, 1, 0)

# Definir la señal original
def x(t):
    return (1/3 * rect(t) +
            2/3 * rect(t - 0.5) +
            1/3 * rect(t - 1.5))

# Definir la señal invertida
def x_inv(t):
    return (1/3 * rect(-t) +
            2/3 * rect(-t - 0.5) +
            1/3 * rect(-t - 1.5))

# Definir la señal invertida y desplazada
def x_inv_shift(t, tau):
    return (1/3 * rect(-t - tau) +
            2/3 * rect(-t - tau - 0.5) +
            1/3 * rect(-t - tau - 1.5))

# Crear el rango de tiempo
t = np.linspace(-2, 2, 1000)

# Graficar las señales
plt.figure(figsize=(12, 8))

# Señal original
plt.subplot(3, 1, 1)
plt.plot(t, x(t), label='x(t)')
plt.title('Señal Original x(t)')
plt.xlabel('t')
plt.ylabel('Amplitud')
plt.grid(True)
plt.legend()

# Señal invertida
plt.subplot(3, 1, 2)
plt.plot(t, x_inv(t), label='x(-t)', color='orange')
plt.title('Señal Invertida x(-t)')
plt.xlabel('t')
plt.ylabel('Amplitud')
plt.grid(True)
plt.legend()

# Señal invertida y desplazada
tau = 0.5
plt.subplot(3, 1, 3)
plt.plot(t, x_inv_shift(t, tau), label='x(-t - τ)', color='red')
plt.title(f'Señal Invertida y Desplazada x(-t - {tau})')
plt.xlabel('t')
plt.ylabel('Amplitud')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
