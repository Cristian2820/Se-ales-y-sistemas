Pruebe otros valores de omega y sigma para crear señales con amplitud creciente/decreciente/constante y diferentes frecuencias angulares.

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

# Definir las variables simbólicas
t, omega, sigma = sym.symbols('t omega sigma')

# Definir la señal e^(st)
s = sigma + sym.I*omega
x = sym.exp(s*t)

# Función para evaluar la señal con valores específicos de omega y sigma
def evaluate_signal(omega_val, sigma_val, t_vals):
    y = x.subs({omega: omega_val, sigma: sigma_val})
    y_func = sym.lambdify(t, y, 'numpy')
    y_vals = y_func(t_vals)
    return y_vals

# Definir el rango de tiempo
t_vals = np.linspace(0, 2*np.pi, 500)

# Configuración 1: Creciente, omega=3, sigma=0.5
omega_val = 3
sigma_val = 0.5
y_vals = evaluate_signal(omega_val, sigma_val, t_vals)

# Graficar la señal
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t_vals, np.real(y_vals), label=f'Re{{e^{{({sigma_val} + j{omega_val})t}}}}')
plt.ylabel('Re{e^{st}}')
plt.ylim([-20, 20])
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t_vals, np.imag(y_vals), label=f'Im{{e^{{({sigma_val} + j{omega_val})t}}}}')
plt.ylabel('Im{e^{st}}')
plt.ylim([-20, 20])
plt.legend()

plt.xlabel('Tiempo [s]')
plt.show()
```

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

# Definir las variables simbólicas
t, omega, sigma = sym.symbols('t omega sigma')

# Definir la señal e^(st)
s = sigma + sym.I*omega
x = sym.exp(s*t)

# Función para evaluar la señal con valores específicos de omega y sigma
def evaluate_signal(omega_val, sigma_val, t_vals):
    y = x.subs({omega: omega_val, sigma: sigma_val})
    y_func = sym.lambdify(t, y, 'numpy')
    y_vals = y_func(t_vals)
    return y_vals

# Definir el rango de tiempo
t_vals = np.linspace(0, 2*np.pi, 500)

# Diferentes configuraciones de omega y sigma
configurations = [
    {'omega': 3, 'sigma': 0.5, 'label': 'Creciente, omega=3, sigma=0.5'},
    {'omega': 3, 'sigma': -0.5, 'label': 'Decreciente, omega=3, sigma=-0.5'},
    {'omega': 3, 'sigma': 0, 'label': 'Constante, omega=3, sigma=0'},
    {'omega': 1, 'sigma': 0.5, 'label': 'Creciente, omega=1, sigma=0.5'},
    {'omega': 5, 'sigma': 0.5, 'label': 'Creciente, omega=5, sigma=0.5'},
]

# Graficar las señales
plt.figure(figsize=(12, 10))

for config in configurations:
    y_vals = evaluate_signal(config['omega'], config['sigma'], t_vals)
    plt.subplot(2, 1, 1)
    plt.plot(t_vals, np.real(y_vals), label=config['label'])
    plt.ylabel('Re{e^{st}}')
    plt.ylim([-20, 20])
    plt.legend()

    plt.subplot(2, 1, 2)
    plt.plot(t_vals, np.imag(y_vals), label=config['label'])
    plt.ylabel('Im{e^{st}}')
    plt.ylim([-20, 20])
    plt.legend()

plt.xlabel('Tiempo [s]')
plt.show()
```

Compruebe la propiedad de selectividad, para un impulso de Dirac desplazado $\delta(t-\tau)$ y probar los resultados modificando el ejemplo anterior.

```python
import sympy as sym
from IPython.display import display, Math

# Definir las variables simbólicas
t, tau = sym.symbols('t tau')

# Definir el impulso de Dirac desplazado
delta_shifted = sym.DiracDelta(t - tau)

# Definir la función genérica f(t)
f = sym.Function('f')(t)

# Calcular la integral para comprobar la propiedad de selectividad
result = sym.integrate(f * delta_shifted, (t, -sym.oo, sym.oo))
# Mostrar el resultado en formato LaTeX
display(Math(r'\int_{-\infty}^{\infty} f(t) \delta(t - \tau) \, dt = f(\tau)'))

# Definir la función específica f(t) = t^2
f_specific = t**2

# Calcular la integral con la función específica
result_specific = sym.integrate(f_specific * delta_shifted, (t, -sym.oo, sym.oo))
# Mostrar el resultado en formato LaTeX
display(Math(r'\int_{-\infty}^{\infty} t^2 \delta(t - \tau) \, dt = \tau^2'))
```

Use la función $\text{rect}(t)$ para construir una función $\cos(\omega t)$ con $\omega=2$, la cual se enciende en $t=-\frac{1}{2}$ y se apaga en $t=+\frac{1}{2}$.

```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np

# Definir las variables
t = sym.Symbol('t')

# Definir la función de onda senoidal
omega = 2 * np.pi  # Frecuencia de la onda
sine_wave = sym.sin(omega * t)

# Definir la función signo
sign_function = sym.sign(sine_wave)

# Construir la función final del rectificador de onda completa
rectified_wave = sign_function * sine_wave

# Convertir la función a una función NumPy para graficar
func = sym.lambdify(t, rectified_wave, 'numpy')

# Definir el rango de tiempo
time_values = np.linspace(-1, 1, 1000)
signal_values = func(time_values)

# Graficar la onda original y la onda rectificada
plt.figure(figsize=(12, 6))
plt.plot(time_values, np.sin(omega * time_values), label='Onda Senoidal Original')
plt.plot(time_values, signal_values, label='Onda Rectificada (Onda Completa)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Rectificador de Onda Completa')
plt.legend()
plt.grid(True)
plt.show()
```
