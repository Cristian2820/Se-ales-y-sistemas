Graficar la potencia media de una impedancia  $R+j\omega L$ , cuándo es excitada por una fuente sinusoidal.



```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
R = 10  # Resistencia en ohmios
L = 0.1  # Inductancia en henrios
V0 = 1  # Voltaje pico de la fuente sinusoidal

# Rango de frecuencias
omega = np.linspace(0.1, 100, 1000)

# Potencia media para Z = R + jωL
P_RL = 0.5 * (V0**2 * R) / (R**2 + (omega * L)**2)

# Gráfico
plt.figure(figsize=(10, 6))
plt.plot(omega, P_RL, label='Z = R + jωL')
plt.xlabel('Frecuencia angular ω (rad/s)')
plt.ylabel('Potencia media P (W)')
plt.title('Potencia media de Z = R + jωL excitada por una fuente sinusoidal')
plt.legend()
plt.grid(True)
plt.show()
```

Graficar la potencia media de una impedancia  $R−j\frac{1}{\omega C}$ , cuándo es excitada por una fuente sinusoidal.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
R = 10  # Resistencia en ohmios
C = 0.001  # Capacitancia en faradios
V0 = 1  # Voltaje pico de la fuente sinusoidal

# Rango de frecuencias
omega = np.linspace(0.1, 100, 1000)

# Potencia media para Z = R - j(1/(ωC))
P_RC = 0.5 * (V0**2 * R) / (R**2 + (1 / (omega * C))**2)

# Gráfico
plt.figure(figsize=(10, 6))
plt.plot(omega, P_RC, label='Z = R - j(1/(ωC))')
plt.xlabel('Frecuencia angular ω (rad/s)')
plt.ylabel('Potencia media P (W)')
plt.title('Potencia media de Z = R - j(1/(ωC)) excitada por una fuente sinusoidal')
plt.legend()
plt.grid(True)
plt.show()
```
