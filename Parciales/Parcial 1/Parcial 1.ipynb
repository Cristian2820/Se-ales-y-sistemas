# Señal con frecuencia dada en el ejercicio

Este código se define una señal continua x(t) compuesta por tres componentes sinusoidales con diferentes frecuencias y amplitudes. Luego, se muestrea esta señal continua a una frecuencia de muestreo de 5 kHz para obtener una versión discreta x[n]. Se grafican ambas señales en subgráficos separados, mostrando la señal continua en un gráfico de líneas y la señal discreta en un gráfico de tallos. El objetivo es ilustrar el proceso de muestreo de una señal continua y cómo se representa en el dominio discreto

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la señal continua x(t)
def x(t):
    return 3 * np.cos(1000*np.pi*t) + 5 * np.sin(2000*np.pi*t) + 10 * np.cos(11000*np.pi*t)

# Frecuencia de muestreo
fs = 5000  # 5 kHz

# Duración de la señal continua
duración = 0.02  # 20 ms

# Crear el eje de tiempo para la señal continua
t_continuo = np.arange(0, duración, 1/fs)

# Muestrear la señal continua
x_discreto = x(t_continuo)

# Crear el eje de tiempo para la señal discreta
t_discreto = np.arange(0, duración, 1/fs)

# Graficar la señal continua y la señal discreta
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t_continuo, x(t_continuo))
plt.title('Señal Continua x(t)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(t_discreto, x_discreto)
plt.title('Señal Discreta x[n]')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Observación:** Al analizar el teorema de Nyquist en este contexto, se encontró que no se cumple debido a que al considerar la frecuencia más rápida de la señal continua, que es de 11000 Hz, se observó que esta frecuencia es mucho mayor que la frecuencia de muestreo, que es de 5000 Hz. Según el teorema de Nyquist, para evitar el aliasing (superposición de frecuencias), la frecuencia de muestreo debe ser al menos el doble de la frecuencia máxima presente en la señal continua. En este caso, la frecuencia de muestreo no cumple con esta condición, lo que puede resultar en distorsión o pérdida de información al reconstruir la señal continua a partir de la señal discreta.



# Señal con frecuencia que cumpla Nyquist

En este código, se realizó un cambio en la frecuencia de muestreo (fs) respecto al ejercicio anterior. Se aumentó la frecuencia de muestreo a 12000 Hz para cumplir con el teorema de Nyquist, ya que la frecuencia más alta en la señal continua es de 11000 Hz.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la señal continua x(t)
def x(t):
    return 3 * np.cos(1000*np.pi*t) + 5 * np.sin(2000*np.pi*t) + 10 * np.cos(11000*np.pi*t)

# Frecuencia de muestreo
fs = 12000  # 6 kHz

# Duración de la señal continua
duracion = 0.01  # 10 ms

# Crear el eje de tiempo para la señal continua
t_continuo = np.arange(0, duracion, 1/fs)

# Muestrear la señal continua
x_discreto = x(t_continuo)

# Crear el eje de tiempo para la señal discreta
t_discreto = np.arange(0, duracion, 1/fs)

# Graficar la señal continua y la señal discreta
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t_continuo, x(t_continuo))
plt.title('Señal Continua x(t)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(t_discreto, x_discreto)
plt.title('Señal Discreta x[n]')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Observación:** En conclusión, al aumentar la frecuencia de muestreo a 12000 Hz en comparación con el ejercicio anterior, se cumplió con el teorema de Nyquist y se evitó el aliasing en la señal discreta. Esto asegura que la información de la señal continua, con componentes de alta frecuencia, se capture de manera adecuada en la señal discreta. Por lo tanto, al reconstruir la señal continua a partir de la señal discreta, se conservará la información original sin distorsiones significativas, lo que se refleja en las representaciones gráficas de ambas señales.

\begin{align}

Teniendo presente todo lo anterior se  empieza utilizando este código con el fin de calcular una aproximación de la salida de un sistema lineal e invariante en el tiempo mediante el uso de una pendiente y un cero. Primero, se determina la pendiente m de una recta que pasa por dos puntos definidos por xhmax y xhmin en función de los valores máximos y mínimos de la entrada x. Luego, se calcula la intersección en y b de esta recta para asegurar que pase por los puntos mencionados. Finalmente, la salida xh se estima como una recta con la pendiente y la intersección calculadas. Este enfoque simplificado asume que la relación entre la entrada y la salida del sistema puede aproximarse mediante una línea recta en el dominio de interés.

```python
def my_ceropend(x,xhmax=0.020,xhmin=0.004):
  #cálculo de pendiente y cero
  m = (xhmax-xhmin)/(x.max()-x.min())
  b = xhmin - m*x.min()
  xh = m*x+b
  return xh
```

Este código calcula dos señales coseno con frecuencias $F1 = \frac{1}{T1}$ Y $F2 = \frac{1}{T2}$, donde T1=6π y T2=8π son los períodos de las señales. Luego, se crea una señal cuasiperiódica x como la suma de estos dos cosenos con una amplitud de 20. La frecuencia de muestreo Fs se establece en 10 veces la máxima frecuencia entre F1 y F2, y se genera un vector de tiempo t que abarca 3 períodos de la señal cuasiperiódica. Finalmente, se aplica una función my_ceropend para calcular una salida aproximada del sistema para la señal x utilizando los valores xhmax y xhmin.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

T1 = 6*np.pi #perido señal 1
T2 = 8*np.pi #perido señal 2
F1 = 1/T1
F2 = 1/T2
T = 24*np.pi #perido cuasiperiódica
Fs = 10*max(F1,F2) #frecuencia muestreo
t = np.arange(0,3*T,1/Fs) #vector de tiempo, se muestran 3 periodos
x = 20*(np.cos(t/3)+np.cos(t/4))
xhmax = 0.020
xhmin = 0.004
xh = my_ceropend(x,xhmax, xhmin).reshape(-1,1)#aplicar cero y pendiente de 0 a 10
#asegurar vector columna
```

Implementando la cuantización de la señal aproximada xh obtenida previamente, utilizando nbits bits para la cuantización.

Primero, se calcula la cantidad de estados posibles (Ne) y la resolución de la cuantización (res). Luego, se crea un vector de estados (ve) que representa los niveles de cuantización posibles.

Posteriormente, se inicializa un vector xq para almacenar la señal cuantizada. En el bucle for, para cada muestra de la señal aproximada xh, se encuentra el estado más cercano en el vector de estados (ve)*texto en cursiva* utilizando la distancia euclidiana. Este estado encontrado se asigna como el valor cuantizado correspondiente en la señal cuantizada xq.

```python
#cuantizar
nbits = 4 #número bits
Ne = 2**nbits #cantidad de estados
res = (xhmax-xhmin)/Ne #resolución cuantización
ve = np.arange(xhmin,xhmax,res).reshape(-1,1)#vector de estados
#asegurar vector columna

xq = np.zeros(len(xh)) #señal cuantizada
for i in range(len(xh)): #recorrer señal cero y pendiente
    xq[i] = ve[np.argmin(cdist(xh[i].reshape(-1,1),ve))] #se cálcula la distancia desde cada punto al vector de estados
```

Esn los gráficos que se muestran a continuación se reflejan las señales originales y cuantizadas en función del tiempo y las muestras respectivamente, luego de realizar los cálculos de cuantización.

El primer gráfico muestra la señal original x(t) en rojo. Esta señal fue generada como la suma de dos cosenos con diferentes frecuencias y una amplitud de 20.
El segundo gráfico muestra la señal aproximada x(t) en azul y la señal cuantizada xq[n] en amarillo. La señal cuantizada se obtiene aplicando la cuantización al resultado de la aproximación xh usando el número de bits especificado.

Estas gráficas permiten visualizar la señal original y cómo se ve afectada por el proceso de cuantización en el dominio del tiempo y de las muestras.

```python
#gráficas de señales
plt.plot(t,x,'r',label='x(t)')
plt.xlabel('t [s]')
plt.ylabel('Amplitud')
plt.show()
plt.plot(t,xh,'b',label='x(t)')
plt.stem(t,xq,'y',label='xq[n]')
plt.xlabel('n [muestra]')
plt.ylabel('Amplitud')
plt.legend()
plt.show()
```

Este código calcula la respuesta al impulso de un sistema discreto a partir de su respuesta al escalón. Primero, se define la respuesta al escalón del sistema como he = np.array([2,4,1,5,0,10]). Luego, se calcula la respuesta al impulso h mediante la función np.diff(he), que calcula la diferencia entre elementos consecutivos de he. Finalmente, se imprime el resultado de h.

```python
he = np.array([2,4,1,5,0,10]) #respuesta al escalón del slit
h = np.diff(he)#respuesta al impulso del slit
print(h)
```

En este apartado se calcula la convolución discreta entre la señal cuantizada xq y la respuesta al impulso h de un sistema discreto. La convolución se realiza utilizando la función np.convolve(xq, h, mode='same'), donde mode='same' asegura que la salida tenga la misma longitud que la entrada más larga, lo que en este caso garantiza que la señal resultante (y) tenga la misma longitud que xq.

Una vez calculada la convolución, se procede a graficar la señal resultante y[n] utilizando un gráfico de tallos (plt.stem). Esta representación visualiza cómo la señal cuantizada xq es procesada por el sistema representado por la respuesta al impulso h, lo que resulta en la señal de salida (y). La gráfica muestra las amplitudes de y en función de las muestras n.

```python
y = np.convolve(xq,h,mode='same')#respuesta del slit -> se asegura tamaños iguales con mode=same
#graficar señal
plt.stem(t,y,'b',label='y[n]')
plt.legend()
plt.xlabel('n [muestra]')
plt.ylabel('Amplitud')
plt.show()
```
