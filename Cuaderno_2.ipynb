## Función creciente

```python
%matplotlib inline
# para graficar en el entorno
import sympy as sym #importa la librería sympy
sym.init_printing() #se inicializa para mostrar en pantalla

t, sigma, omega = sym.symbols('t sigma omega', real=True)#definir variables simbólicas
s = sigma + 1j*omega  #sobre carga operador para definir parte imaginaria
x = sym.exp(s*t)#función exponencial simbólica desde sympy
x
```

```python
y = x.subs({omega: 3, sigma : 0.8}) #sustituir variables omega y sigma por valores numéricos
print(r'x(t)=',x)
#se gráfica desde el paquete plot de sympy por ser objetos de variable simbólica
sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}',ylim=[-200,200])
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}',ylim=[-200,200])
```

# Función constante

```python
y = x.subs({omega: 8, sigma : 0}) #sustituir variables omega y sigma por valores numéricos
print(r'x(t)=',x)
#se gráfica desde el paquete plot de sympy por ser objetos de variable simbólica
sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}',ylim=[-2,2])
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}',ylim=[-2,2])
```

# Función decreciente



```python
y = x.subs({omega: 10, sigma : -0.7}) #sustituir variables omega y sigma por valores numéricos
print(r'x(t)=',x)
#se gráfica desde el paquete plot de sympy por ser objetos de variable simbólica
sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}',ylim=[-1,1])
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}',ylim=[-1,1])
```

**Observación:** Se puede observar que con la variación del sigma esto provoca con un numero n>0 se obtiene una funcion creciente, cuando es con n=0 se mantiene constante y por último cuando se presenta 0>n esto da una onda decreciente. No obstante si se modifica el omega si el valor es n>0 se obtienen una mayor frecuencia mientras que con 0>n el resultado es una menor frecuencia.

# Impulso de Dirac - Función delta  δ(t)

```python
delta = sym.DiracDelta(t)#función dirac en sympy
delta
```

```python
f = sym.Function('f')(t) # f(t)
sym.integrate(delta*f, (t, -sym.oo, sym.oo)) # integral con limites desde -inf hasta +inf
```

```python
#f(t-\tau)
#definir tau
tau = sym.symbols('tau', real=True)
a= sym.symbols('a', real=True)
g = sym.Function('g')(t) # g(t)
delta_T = sym.DiracDelta(a*(t-tau)) # delta(t-tau)
delta_T
sym.integrate(delta_T*g, (t, -sym.oo, sym.oo))
```

# Función escalon

```python
class rect(sym.Function): #clase/objeto tipo sympy

    @classmethod #decorador objeto
    def eval(cls, arg):
        return sym.Heaviside(arg + sym.S.Half) - sym.Heaviside(arg - sym.S.Half)
```

```python
x=sym.cos(omega*t)
y=x.subs({omega: 2})
sym.plot(rect(t)*y, (t, -1, 1), ylim=[-0.2, 1.2], ylabel=r'rect$(t)$');
```
